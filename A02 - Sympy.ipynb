**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python
import numpy as np
import sympy as sp

sp.var("x")

def f(x):
    return 1260 + 117*x - 9*x**2

df = sp.diff(f(x), x)

pc = sp.solve(df,x)
pc
```

```python
ddf = sp.diff(df, x)
ddf
```

```python
crit_2da_derivada = ddf.subs(x, pc[0])
crit_2da_derivada
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

xn = np.linspace(-10,10,1000)

plt.figure()
plt.plot(xn, f(xn))
plt.plot(pc[0], f(pc[0]), "*m", ms=10)
```

La función tiene un punto crítico en 13/2.
El punto crítico es un máximo global.

**El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).**

```python
sp.var("x")
sp.var("y")

f = 2*x + 2*y - 10
sol_y = sp.solve(f,y)[0]
sol_y
```

```python
h = (x**2 - sol_y**2)**(1/2)
h2 = (25-10*x)**(1/2)
h2
# h y h2 son lo mismo, solo que h2 está simplificado porque Python no lo simplifica por si solo
```

```python
def v(x):
    return (3.141592 * x**2 * (25-10*x)**(1/2))/3

dv = sp.diff(v(x), x)

pc = sp.solve(dv,x)
for c in pc:
    if c > 0:
        print(c)
```

```python
ddv = sp.diff(dv, x)
ddv
```

```python
for c in pc:
    if c > 0:
        crit_2da_derivada = ddv.subs(x, c)
        print(crit_2da_derivada)
```

```python
xn = np.linspace(0,2.5,1000)

plt.figure()
plt.plot(xn,v(xn))
for c in pc:
    if c > 0:
        plt.plot(c, v(c), "*m", ms=10)
```

Encontramos un punto crítico en x = 2. Este es un máximo global. 
Esto significa que el mayor volumen se alcanza cuando el radio es de 2 cm, es decir, la base del triángulo isóceles es de 4 cm. Esto significa que los otros dos lados miden 3 cm cada uno. Al sustituir estos valores en la fórmula obtenemos un volumen de 9 cm cúbicos.

**Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?**

```python
sp.var("x")
sp.var("y")

f = 2*x + y - 6
sol_y = sp.solve(f,y)[0]
sol_y
```

```python
def a(x):
    return x * (6-2*x)

da = sp.diff(a(x), x)

pc = sp.solve(da,x)
for c in pc:
    if c > 0:
        print(c)
```

```python
dda = sp.diff(da, x)
dda
```

```python
xn = np.linspace(0,3,1000)

plt.figure()
plt.plot(xn,a(xn))
for c in pc:
    if c > 0:
        plt.plot(c, a(c), "*m", ms=10)
```

Existe un punto crítico en x = 3/2. Este es un máximo global. 
Esto significa que el área máxima de la porteria se alcanza cuando los postes miden 1.5 m (3/2 m) y el travesaño mide 3 m. De tal manera se utilizan los 6 metros de la barra de aluminio y se genera un área de 4.5 metros cuadrados.
